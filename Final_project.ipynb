```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#- Index Model : Vector space model
class vsmIndexModel:

    def __init__(self, documents_df):
        """input:
            documents_df (DataFrame): contains the preprocessed documents
            fit
            transfrm
            fit_transform
        """
        # Create a TF-IDF vectorizer
        self.tfidf_vectorizer = TfidfVectorizer()
        #-then finally, fit the documents and transform
        self._index = self.tfidf_vectorizer.fit_transform(documents_df["summary"])

    def getindex(self):
        return self._index

    def vectorize(self, sentence):
        if isinstance(sentence,str):
            qry=pd.DataFrame([{"summary":sentence}])
        else:
            qry=sentence
        return self.tfidf_vectorizer.transform(qry['summary'])
```

```python

class Retriever:

    def retrieve(self, query_vec, index_model):
        cosine_similarities = cosine_similarity(query_vec, index_model.getindex())
        results = pd.DataFrame(
            [{'docno':i, 'score':cosine_similarities[0][i], 'content':None}
            for i in range(len(cosine_similarities[0]))]
        ).sort_values(by=['score'], ascending=False)
        return results[results["score"]>0]
```

```python
class SearchEngine:
    def __init__(self, preprocessor, retriever, documents):
        self.preprocessor = preprocessor
        self.retriever = retriever
        self.documents = None
        self.model = None
        self.rebuild(documents)

    def rebuild(self, documents):
        self.documents = documents
        self.documents['text'] = self.documents['summary'].apply(self.preprocessor.process)
        self.model = vsmIndexModel(self.documents)

    def querying(self, query):
        query_prc = self.preprocessor.process(query)
        query_vec = self.model.vectorize(query_prc)
        docs_res = self.retriever.retrieve(query_vec, self.model)
        if docs_res.shape[0]>0:
            docs_res['content'] = docs_res.apply(
                lambda row: self.documents[row.docno]['title'].iloc[0], axis = 1)
        return docs_res
```

```python
vsm_model = vsmIndexModel(docs_df)

vsm_retriever = Retriever()

topics_df = pd.DataFrame(topics)
vsm_results = []

for query_id, query_text in zip(topics_df['qid'], topics_df['query']):
    query_vector = vsm_model.vectorize(query_text)
    retrieved_docs = vsm_retriever.retrieve(query_vector, vsm_model)
    for _, row in retrieved_docs.iterrows():
        vsm_results.append({'qid': query_id, 'docno': row['docno'], 'score': row['score']})

vsm_results_df = pd.DataFrame(vsm_results)
```


### تحليل النتائج:


1. **قيم "map"**:
   - جميع النماذج تقريباً تعطي قيمة منخفضة جدًا لـ "map"، مما يشير إلى أن الترتيب العام للوثائق المسترجعة ليس جيدًا جدًا فيما يتعلق بمطابقة استعلامات المستخدم مع الوثائق ذات الصلة.
   - نموذج VSM لم يحقق أي نتائج مفيدة (قيمة "map" غير متوفرة - NaN).

2. **قيم "ndcg"**:
   - القيم منخفضة جدًا في جميع النماذج، مما يشير إلى أن الوثائق الأعلى ترتيبًا ليست وثائق مفيدة جدًا. يبدو أن النماذج لا ترتب الوثائق بطريقة تجعل الأكثر صلة في المقدمة.
   - نموذج VSM يظهر قيمة "ndcg" تساوي 0.000، مما يشير إلى أن النموذج لم يسترجع أي وثائق ذات صلة.

3. **قيم Recall **:
   - قيم الاسترجاع عبر عدة مستويات (R@5, R@10, R@20, إلخ) تظهر أن النماذج تفشل في استرجاع الوثائق ذات الصلة. القيم قليلة جدًا مما يعني أن النماذج لم تعثر على معظم الوثائق ذات الصلة بالاستعلامات.
   - في حالة نموذج VSM، القيم كلها صفر، مما يعني أن النموذج لم يسترجع أي وثيقة ذات صلة.

4. **قيم presision **:
   - قيم الدقة في مستويات مختلفة (P@5, P@10, P@20, إلخ) منخفضة جدًا، مما يعني أن نسبة الوثائق المسترجعة التي كانت ذات صلة بالفعل قليلة جدًا.
   - نموذج VSM لم يسترجع أي وثيقة ذات صلة، مما أدى إلى حصوله على قيم صفرية في جميع مستويات الدقة.
